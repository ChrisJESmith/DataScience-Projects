# Visualising Earnings Based on College Majors

The following dataset shows job outcomes of students who graduated from college between 2010 and 2012. The original data was released by the American Community Survey. FiveThirtyEight cleaned the dataset and released it on their Github page.

Each row represents a different major (subject) in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- Rank - rank by median earnings (dataset is ordered by this column)
- Major_code - major code
- Major - major description
- Major_category - category of major
- Total - total number of people with major
- Sample_size - sample size (unweighted) of full-time
- Men - male graduates
- Women - female graduates
- ShareWomen - women as share of total
- Employed - number employed
- Median - median salary of full-time, year-round workers
- Low_wage_jobs - number in low-wage service jobs
- Full_time - number employed 35 hours or more
- Part_time - number employed less than 35 hours

## Pandas Histograms

We will generate a series of histograms to explore the distributions of various columns. A histogram represents the frequency distribution of continuous/numerical variables such as 'Age' (whereas a bar graph is a diagrammatic comparison of discrete variables showing categorical data)

```python
# list of columns
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

# create a fig object which determines the SIZE of the plots
fig = plt.figure(figsize=(5,12))

# for each iterator (1-4)
for r in range(1,5):
    
    # create a series of plots with the parameters of 4 (graph spacing), 
    # 1 (one plot per line i.e. stacked on top of each other), 
    # and r (the plot itself i.e. 4 separate graphs)
    ax = fig.add_subplot(4,1,r)
    
    # once the plot layout is defined, plot the actual data
    # use the reference to the columns in 'cols' to plot the data for each
    # based on the index (1-4) this plots "Median", "Employed", "Full_time", "ShareWomen" columns
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

```python
# create the same plots as above, but plot the last four columns in the list
fig = plt.figure(figsize=(5,12))

for r in range(4,8):
    
    # need to write 'r-3' as index for the loop must start at 1, to generate the subplots
    ax = fig.add_subplot(4,1,r-3)
    
    # based on the index (4-8), this plots "ShareWomen", "Unemployment_rate", "Men", "Women" columns
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

## Bar Plots

We will generate some bar plots, which are used to compare categorical variables (whereas histograms plot continuous data with ranges of the data grouped into bins or intervals)

```python
# compare the percentages of women against majors (subjects) - based on the first ten rows
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
# compare the percentages of women against majors (subjects) - based on the last ten rows
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

## Findings 

We can see a greater proportion of women do jobs that are ranked at the bottom end of the median salary scale

```python
# compare the unemployment rate against majors (subjects) - based on the first ten rows
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

```python
# compare the unemployment rate against majors (subjects) - based on the bottom ten rows
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

## Findings 

There doesn't seem to be any clear difference between the unemployment rate and subjects ranked as the highest and lowest on the median salary scale
